```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load all tables
customers = pd.read_csv('/content/olist_customers_dataset.csv')
orders = pd.read_csv('/content/olist_orders_dataset.csv')
order_items = pd.read_csv('/content/olist_order_items_dataset.csv')
products = pd.read_csv('/content/olist_products_dataset.csv')
payments = pd.read_csv('/content/olist_order_payments_dataset.csv')

# First look at each table
for name, df in [('Customers', customers),
                 ('Orders', orders),
                 ('Order Items', order_items),
                 ('Products', products),
                 ('Payments', payments)]:
    print(f"\n{name}:")
    print(f"  Rows: {len(df):,}")
    print(f"  Columns: {df.columns.tolist()}")
```

```python
# Check missing values in each table
for name, df in [('Customers', customers),
                 ('Orders', orders),
                 ('Order Items', order_items),
                 ('Products', products),
                 ('Payments', payments)]:
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    print(f"\n{name} — Missing values:")
    if len(missing) == 0:
        print("  No missing values ✅")
    else:
        print(missing)
```

```python
# Check order status distribution first
print("Order Status breakdown:")
print(orders['order_status'].value_counts())
```

```python
# Keep only delivered orders
orders_delivered = orders[
    orders['order_status'] == 'delivered'
].copy()

# Convert date columns to datetime
date_cols = ['order_purchase_timestamp',
             'order_approved_at',
             'order_delivered_carrier_date',
             'order_delivered_customer_date',
             'order_estimated_delivery_date']

for col in date_cols:
    orders_delivered[col] = pd.to_datetime(
        orders_delivered[col]
    )

print(f"Delivered orders: {len(orders_delivered):,}")
print(f"Dropped: {len(orders) - len(orders_delivered):,} non-delivered orders")
print(f"\nDate range:")
print(f"First order: {orders_delivered['order_purchase_timestamp'].min().date()}")
print(f"Last order: {orders_delivered['order_purchase_timestamp'].max().date()}")
```

```python
# Step 1 — Merge orders + customers
df = orders_delivered.merge(customers,
                            on='customer_id',
                            how='left')

# Step 2 — Merge with payments
payments_agg = payments.groupby('order_id').agg(
    total_payment=('payment_value', 'sum'),
    payment_type=('payment_type', 'first'),
    installments=('payment_installments', 'max')
).reset_index()

df = df.merge(payments_agg,
              on='order_id',
              how='left')

# Step 3 — Merge with order items
items_agg = order_items.groupby('order_id').agg(
    total_items=('order_item_id', 'count'),
    total_price=('price', 'sum'),
    freight_value=('freight_value', 'sum')
).reset_index()

df = df.merge(items_agg,
              on='order_id',
              how='left')

# Step 4 — Merge with products (most common category per order)
order_products = order_items.merge(
    products[['product_id', 'product_category_name']],
    on='product_id',
    how='left'
)
category_per_order = order_products.groupby('order_id')[
    'product_category_name'
].first().reset_index()
category_per_order.columns = ['order_id', 'category']

df = df.merge(category_per_order,
              on='order_id',
              how='left')

# Step 5 — Add useful time features
df['purchase_month'] = df[
    'order_purchase_timestamp'
].dt.to_period('M')
df['purchase_year'] = df[
    'order_purchase_timestamp'
].dt.year
df['delivery_days'] = (
    df['order_delivered_customer_date'] -
    df['order_purchase_timestamp']
).dt.days
df['delivery_vs_estimate'] = (
    df['order_delivered_customer_date'] -
    df['order_estimated_delivery_date']
).dt.days  # negative = early, positive = late

print(f"Master table shape: {df.shape}")
print(f"\nColumns: {df.columns.tolist()}")
print(f"\nSample:")
df.head(3)
```

```python
# Delivery Performance Analysis
print("=== DELIVERY PERFORMANCE ===\n")

# Average delivery time
print(f"Avg delivery days: {df['delivery_days'].mean():.1f} days")
print(f"Fastest delivery: {df['delivery_days'].min()} days")
print(f"Slowest delivery: {df['delivery_days'].max()} days")

# Early vs late deliveries
early = (df['delivery_vs_estimate'] < 0).sum()
on_time = (df['delivery_vs_estimate'] == 0).sum()
late = (df['delivery_vs_estimate'] > 0).sum()
total = len(df)

print(f"\nDelivery Promise Accuracy:")
print(f"Early deliveries: {early:,} ({early/total*100:.1f}%)")
print(f"On time: {on_time:,} ({on_time/total*100:.1f}%)")
print(f"Late deliveries: {late:,} ({late/total*100:.1f}%)")

# Average days early or late
print(f"\nAvg days early: {abs(df[df['delivery_vs_estimate']<0]['delivery_vs_estimate'].mean()):.1f} days")
print(f"Avg days late: {df[df['delivery_vs_estimate']>0]['delivery_vs_estimate'].mean():.1f} days")

# Monthly revenue trend
print(f"\n=== REVENUE OVERVIEW ===")
print(f"Total Revenue: R$ {df['total_payment'].sum():,.2f}")
print(f"Avg Order Value: R$ {df['total_payment'].mean():.2f}")
print(f"Total Orders: {len(df):,}")
print(f"Total Customers: {df['customer_unique_id'].nunique():,}")
```

```python
# Retention Analysis
print("=== CUSTOMER RETENTION ANALYSIS ===\n")

# Orders per customer
orders_per_customer = df.groupby(
    'customer_unique_id'
)['order_id'].count()

print("Purchase frequency distribution:")
print(orders_per_customer.value_counts()
      .head(5).to_string())

one_time = (orders_per_customer == 1).sum()
repeat = (orders_per_customer > 1).sum()
total_customers = len(orders_per_customer)

print(f"\nOne-time buyers: {one_time:,} ({one_time/total_customers*100:.1f}%)")
print(f"Repeat buyers: {repeat:,} ({repeat/total_customers*100:.1f}%)")
print(f"\nTop customer orders: {orders_per_customer.max()}")
print(f"Avg orders per customer: {orders_per_customer.mean():.2f}")

# Revenue from repeat vs one-time
customer_revenue = df.groupby(
    'customer_unique_id'
)['total_payment'].sum()
repeat_customers = orders_per_customer[
    orders_per_customer > 1
].index
one_time_customers = orders_per_customer[
    orders_per_customer == 1
].index

repeat_revenue = customer_revenue[repeat_customers].sum()
onetime_revenue = customer_revenue[one_time_customers].sum()
total_revenue = customer_revenue.sum()

print(f"\n=== REVENUE SPLIT ===")
print(f"Revenue from repeat buyers: R$ {repeat_revenue:,.2f} ({repeat_revenue/total_revenue*100:.1f}%)")
print(f"Revenue from one-time buyers: R$ {onetime_revenue:,.2f} ({onetime_revenue/total_revenue*100:.1f}%)")
```

```python
# RFM Analysis
import datetime

print("=== RFM SEGMENTATION ===\n")

# Reference date = day after last order
reference_date = df['order_purchase_timestamp'].max() + \
                 datetime.timedelta(days=1)

# Calculate RFM for each customer
rfm = df.groupby('customer_unique_id').agg(
    recency=('order_purchase_timestamp',
             lambda x: (reference_date - x.max()).days),
    frequency=('order_id', 'count'),
    monetary=('total_payment', 'sum')
).reset_index()

print("RFM Table Sample:")
print(rfm.head())
print(f"\nRFM Shape: {rfm.shape}")
print(f"\nRecency stats (days since last purchase):")
print(f"  Min: {rfm['recency'].min()} days")
print(f"  Max: {rfm['recency'].max()} days")
print(f"  Avg: {rfm['recency'].mean():.0f} days")
print(f"\nMonetary stats:")
print(f"  Min spend: R$ {rfm['monetary'].min():.2f}")
print(f"  Max spend: R$ {rfm['monetary'].max():.2f}")
print(f"  Avg spend: R$ {rfm['monetary'].mean():.2f}")
```

```python
# Score each customer 1-5 on R, F, M
# Recency: lower days = higher score (5 is best)
# Frequency: higher count = higher score
# Monetary: higher spend = higher score

rfm['R_score'] = pd.qcut(rfm['recency'],
                          q=5,
                          labels=[5,4,3,2,1])

rfm['F_score'] = pd.qcut(rfm['frequency'].rank(
                          method='first'),
                          q=5,
                          labels=[1,2,3,4,5])

rfm['M_score'] = pd.qcut(rfm['monetary'],
                          q=5,
                          labels=[1,2,3,4,5])

# Convert to int
rfm['R_score'] = rfm['R_score'].astype(int)
rfm['F_score'] = rfm['F_score'].astype(int)
rfm['M_score'] = rfm['M_score'].astype(int)

# Combined RFM score
rfm['RFM_score'] = rfm['R_score'].astype(str) + \
                   rfm['F_score'].astype(str) + \
                   rfm['M_score'].astype(str)

# Segment customers
def segment_customer(row):
    r = row['R_score']
    f = row['F_score']
    m = row['M_score']

    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'
    elif r >= 3 and f >= 3:
        return 'Loyal'
    elif r >= 4 and f <= 2:
        return 'New Customers'
    elif r <= 2 and f >= 3:
        return 'At Risk'
    elif r <= 2 and f <= 2 and m <= 2:
        return 'Lost'
    else:
        return 'Potential Loyalists'

rfm['segment'] = rfm.apply(segment_customer, axis=1)

# Results
print("=== CUSTOMER SEGMENTS ===\n")
segment_summary = rfm.groupby('segment').agg(
    customer_count=('customer_unique_id', 'count'),
    avg_recency=('recency', 'mean'),
    avg_frequency=('frequency', 'mean'),
    avg_monetary=('monetary', 'mean')
).round(1)

segment_summary['revenue_contribution'] = (
    rfm.groupby('segment')['monetary'].sum() /
    rfm['monetary'].sum() * 100
).round(1)

print(segment_summary.sort_values(
    'customer_count', ascending=False))
```

```python
# Business priority summary
print("=== BUSINESS PRIORITY MATRIX ===\n")

total_customers = rfm['customer_unique_id'].count()
total_revenue = rfm['monetary'].sum()

for segment in ['Champions', 'At Risk',
                'Potential Loyalists', 'Loyal',
                'New Customers', 'Lost']:
    seg_data = rfm[rfm['segment'] == segment]
    count = len(seg_data)
    revenue = seg_data['monetary'].sum()
    avg_days = seg_data['recency'].mean()

    print(f"{segment}:")
    print(f"  Customers : {count:,} ({count/total_customers*100:.1f}%)")
    print(f"  Revenue   : R$ {revenue:,.0f} ({revenue/total_revenue*100:.1f}%)")
    print(f"  Avg days since purchase: {avg_days:.0f}")
    print()

# Save for Power BI
df_rfm = df.merge(
    rfm[['customer_unique_id',
         'R_score', 'F_score', 'M_score',
         'RFM_score', 'segment',
         'recency', 'frequency', 'monetary']],
    on='customer_unique_id',
    how='left'
)

df_rfm.to_csv('/content/ecommerce_powerbi.csv',
              index=False)
print("✅ Saved ecommerce_powerbi.csv — ready for Power BI!")
print(f"Shape: {df_rfm.shape}")
```
